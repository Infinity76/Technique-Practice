```python
import pandas as pd
import numpy as np
import bamboolib as bam
```

```python
df = pd.read_csv("bike_crash (1).csv")
```

```python
df
```

```python
df2 = df
```

```python
Crash_mapping = {
           'Non-Incapacitating Injury': (0),
           "Possible Injury":(0),
           "Incapacitating Injury": (1),
           'Not Injured': (0),
           'Killed': (1)}


df2["Crash Severity"] = df2["Crash Severity"].map(Crash_mapping)
```

```python
df2
```

```python
df2["$1000 Damage to Any One Person's Property"] = pd.get_dummies(df2["$1000 Damage to Any One Person's Property"], drop_first=True)
```

```python
df2
```

```python
df2["At Intersection Flag"] = pd.get_dummies(df["At Intersection Flag"],drop_first=True)
```

```python
df3 = df2
df3
```

```python
df_week = pd.get_dummies(df3["Day of Week"])
```

```python
df_week
```

```python
df_Intersection_Related = pd.get_dummies(df3["Intersection Related"])
```

```python
df_Intersection_Related = df_Intersection_Related.drop(columns=['Not Reported'])
df_Intersection_Related
```

```python
df_worn_helmet = pd.get_dummies(df["Person Helmet"])
```

```python
df_worn_helmet
```

```python
df_Traffic_Control_Type = pd.get_dummies(df["Traffic Control Type"]) 
```

```python
df_Traffic_Control_Type
```

```python
df5 = df3[["$1000 Damage to Any One Person's Property","At Intersection Flag","Crash Time","Crash Total Injury Count","Crash Year","Speed Limit"]]
```

```python
X_df = [df_worn_helmet,df_week,df_Traffic_Control_Type,df_Intersection_Related,df5]
```

```python
X = pd.concat(X_df)
```

```python
X
```

```python
X = pd.concat([df_week, df_worn_helmet,df5, df_Intersection_Related,df_Traffic_Control_Type], axis=1, join='inner')
y = df2["Crash Severity"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.25)
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
Prediction = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
classification_report(y_test,Prediction)
```

```python
cm_LR = confusion_matrix(y_test,Prediction)
cm_LR
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as ply
plt.figure(figsize=(3,3))
sns.heatmap(cm_LR,annot=True)
plt.xlabel('Predictor')
plt.ylabel("Truth")

```
